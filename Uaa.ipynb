```python

import numpy as np
from sklearn import preprocessing

import seaborn as sns

#Напишем функцию, которая принимает на вход наши данные, кодирует числовыми значениями категориальные признаки
#и возвращает обновленный данные и сами кодировщики
def number_encode_features(init_df):
    result = init_df.copy() #копируем нашу исходную таблицу
    encoders = {}
    for column in result.columns:
        if result.dtypes[column] == object: # np.object -- строковый тип / если тип столбца - строка, то нужно его закодировать
            encoders[column] = preprocessing.LabelEncoder() #для колонки column создаем кодировщик
            result[column] = encoders[column].fit_transform(result[column]) #применяем кодировщик к столбцу и перезаписываем столбец
    return result, encoders

encoded_data, encoders = number_encode_features(df) #Теперь encoded data содержит закодированные категориальные признаки
encoded_data.head() #проверяем
#
```

```python
#смотрим есть ли линейная зависимоть у таргета

plt.subplots(figsize=(10,10))
encoded_data, encoders = number_encode_features(df)
sns.heatmap(encoded_data.corr(), square=True, annot=True) #можем заметить, что есть пересечения с ценой билета и полом, а с возрастом почти нет пересечений
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Предположим, что ваш DataFrame называется df и содержит колонки 'floors_count' и 'floor'

# Извлекаем данные по этажам
floors_count_data = df['floors_count'].dropna()
floor_data = df['floor'].dropna()

# Определяем минимальные и максимальные значения
xmin = min(floors_count_data.min(), floor_data.min())
xmax = max(floors_count_data.max(), floor_data.max())
width = (xmax - xmin) / 20  # Настройка ширины

# Строим наложенные гистограммы
plt.figure(figsize=(10, 6))
sns.histplot(floors_count_data, color='b', bins=np.arange(xmin, xmax, width), kde=False, alpha=0.5, label='floors_count')
sns.histplot(floor_data, color='r', bins=np.arange(xmin, xmax, width), kde=False, alpha=0.5, label='floor')

plt.title('Наложенные гистограммы для этажей')
plt.xlabel('Этажи')
plt.ylabel('Количество')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Предположим, что ваш DataFrame называется df и содержит колонки 'price', 'rooms_count', 'total_meters', 'floors_count', 'floor'

# Создаем фигуру с подграфиками
plt.figure(figsize=(15, 10))

# Подграфик для rooms_count
plt.subplot(221)
sns.scatterplot(data=df, x='rooms_count', y='price')
plt.title('Price vs Rooms Count')
plt.xlabel('Rooms Count')
plt.ylabel('Price')

# Подграфик для total_meters
plt.subplot(222)
sns.scatterplot(data=df, x='total_meters', y='price')
plt.title('Price vs Total Meters')
plt.xlabel('Total Meters')
plt.ylabel('Price')

# Подграфик для floors_count
plt.subplot(223)
sns.scatterplot(data=df, x='floors_count', y='price')
plt.title('Price vs Floors Count')
plt.xlabel('Floors Count')
plt.ylabel('Price')

# Подграфик для floor
plt.subplot(224)
sns.scatterplot(data=df, x='floor', y='price')
plt.title('Price vs Floor')
plt.xlabel('Floor')
plt.ylabel('Price')

# Общий заголовок
plt.suptitle('Влияние Price на различные характеристики', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Для корректного отображения заголовка
plt.show()
```
